## Real Estate Investment Analysis

The goal of this analysis is to identify attractive property investments based on financial metrics such as rental yield, price-to-rent ratio, and vacancy rates.


### Dataset

We generate a small dataset of properties including info about:
- Prices
- Rents
- Sizes
- Vacancy rates

```python
prices = [180000, 275000, 320000, 450000, 220000, 510000]
rents = [950, 1400, 1600, 2100, 1200, 2600]
sizes = [55, 70, 90, 110, 65, 130]
vacancy_rates = [0.05, 0.08, 0.03, 0.12, 0.07, 0.02]
```

### Data Cleaning
We remove the invalid observations from our dataset where the price or rent is negative and where the vacancy rate exceeds 20%. 

```python
cleaned_data = [(p,r,s,v) for (p,r,s,v) in zip(prices, rents, sizes, vacancy_rates) if p >0 and r >0 and v <= 0.2]
cleaned_data
```

### Feature Engineering
We compute some financial metrics which we will use later on in our analysis to filter on. The metrics include:
- **Price-to-rent ratio**
- **Yield adjusted by vacancy rate**

```python
ratios = [p/(12*r) for (p,r,s,v) in cleaned_data]
yields = [(r*12*(1-v))/ p for (p,r,s,v) in cleaned_data]
```

### Classification
We clasify the properties based on their corresponding adjusted yields:
- Low: < 5%
- Medium: 5-8%
- High: > 8%

```python
labels = ["Low" if y < 0.05 else "Medium" if y <= 0.08 else "High" for (p,r,s,v), y in zip(cleaned_data, yields)]
```

### Investment criteria
We will select the most attractive investments based on some criteria:
- Yield > 6%
- Price-to-rent ratio < 20%
- Vacancy rate < 10%

```python
top_investments = [(p, r, s, v, y, l) for (p, r, s, v), ratio, y, l in zip(cleaned_data, ratios, yields, labels) if y > 0.06 and ratio < 20 and v < 0.10]
top_investments
```
